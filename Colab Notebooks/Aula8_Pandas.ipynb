```python
#Criando nosso DataFrame
df = pd.read_excel("/content/drive/MyDrive/datasets/AdventureWorks.xlsx")
```

```python
#Visualizando as 5 primeiras linhas
df.head()
```

```python
#Quantidae de linhas e colunas
df.shape
```

```python
#Verificando os tipo de dados
df.dtypes
```

```python
#Qual a Receita Total?
df["Valor Venda"].sum()
```

```python
#Qual o Custo Total?
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"]) #Criando a coluna de custo
```

```python
df.head(1)
```

```python
#Qual o Custo Total?
round(df["Custo"].sum(), 2)
```

```python
#Agora temos a Receita e o Custo Total, podemos achar o Lucro Total
#Vamos criar uma coluna de Lucro que será Receita - Custo
df["Lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
df.head(1)
```

```python
#Lucro Total?
round(df["Lucro"].sum(), 2)
```

```python
#Criando uma coluna com o total de dias para enviar o produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

**Agora, queremos saber a média do tempo de envio para cada Marca, e para isso precisamos transformar a coluna Tempo_envio em numérica**

```python
#Extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
#Verificando o tipo da coluna Tempo_envio
df["Tempo_envio"].dtype
```

```python
#Média do tempo de envio por Marca
df.groupby("Marca")["Tempo_envio"].mean()
```

**Missing Values**

```python
#Verificando se temos dados faltantes
df.isnull().sum()
```

**E se a gente quiser saber o Lucro por Ano e por Marca?**

```python
#Vamos agrupar por Ano e Marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
#Formatando os valores numéricos para tirar a notação científica
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Gráfico Total de Produtos Vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total de Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
#Lucro por Ano
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
#Selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
from ctypes import pointer
#Gráfico de linha para Lucro x Mês
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro x Mês", marker = "o")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal'); #Rotaciona as marcações do eixo x para a horizontal
```

```python
#Lucro x Classe
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```
